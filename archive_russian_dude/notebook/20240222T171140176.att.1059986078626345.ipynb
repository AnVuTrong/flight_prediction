```python
import pandas as pd
data_file = './datafiles/Data - Route 6 - 280 flights.xlsx'
data_train = pd.read_excel(data_file,'Sum data- 280c') 
data_train_high =data_train.drop(data_train[data_train.Ac_feet < 13500].index)
data_train_high = data_train_high.replace(regex=['↓', '↘','→','°','↙ ','← ','↗ ','↑ ','↖ '], value='')

data_train_high=data_train_high.fillna(0)
```

```python
import os
import pandas as pd
from lifelines.utils import concordance_index
import seaborn as sns
import pandas as pd
thresh =0.0
targets= ['Ac_Lon','Ac_feet','Ac_kts','Ac_Course','Ac_Lat']
plt.rcParams["figure.figsize"] = (40,20)
for target in targets:
    print('############################################################')
    print('############################################################')
    print('Tinh tuong quan C-Index giua cac tham so doi voi target '+target)

    events= data_train_high[target].tolist()
    feas = list(data_train_high.drop(columns=['Time (MSK)', 'validdate', 'Ac_id','Ac_Lon','Ac_feet','Ac_kts','Ac_Course','new_time','Time_step','Ac_Lat']).columns.values)
    cin=[]
    for fea in feas:
        preds = data_train_high[fea].tolist()
        c_index = concordance_index(events, preds)
        cin.append(c_index)

        print(fea,'Concordance index: ',concordance_index(events, preds))
    data = {'Features' : feas, 'CI' : cin}

    df = pd.DataFrame(data)
    df = df.sort_values(by=['CI'],ascending=False)
    df_highci =df.drop(df[df.CI <=thresh].index)
    df_highci.to_excel('./output_c_index/fea_cindex_'+target+'.xlsx', index=False)
    # #get list of high cindex
    df_fea_high = pd.read_excel('./output_c_index/fea_cindex_'+target+'.xlsx')
    high_ci_fea_list = df_fea_high.Features.tolist()
    print('Tong so feature: ', len(high_ci_fea_list))
    g=sns.lineplot(data=df_fea_high,
        x="Features", y="CI",
        markers=True, dashes=True
    )
    print('############################################################')
    print('############################################################')
plt.legend(title='Corrlation between each parameter to specific targets', loc='upper left', labels=targets)
plt.show(g)
g.figure.savefig('./images/correlation_between_input_with_targets.png')
```
