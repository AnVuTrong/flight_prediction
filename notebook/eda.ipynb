```python
def lst_of_unreasonable_time(df, max_threshold=10000):
    unreasonable_time = df[(df['Time_step'] >= max_threshold)]
    return unreasonable_time

# Example usage
unreasonable_times = lst_of_unreasonable_time(raw)
unreasonable_times
```

```python
def remove_unreasonable_time(df, max_threshold=10000, min_flight_time=4000):
    # Remove rows with Time_step greater than max_threshold
    cleaned_df = df[df['Time_step'] < max_threshold]
    
    # Identify flights with maximum Time_step less than min_flight_time
    flights_to_remove = cleaned_df.groupby('Ac_id')['Time_step'].max() < min_flight_time
    id_to_remove = flights_to_remove[flights_to_remove].index
    
    # Remove these flights from the cleaned dataframe
    cleaned_df = cleaned_df[~cleaned_df['Ac_id'].isin(id_to_remove)]
    
    return cleaned_df

cleaned = remove_unreasonable_time(raw, max_threshold=10000)
cleaned
```

```python
import torch
from modules.models import FlightLSTM
from modules.feature_engineering import FeatureEngineeringV1
fe = FeatureEngineeringV1()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
csv_path = '../data/csv/raw.csv'
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df_normalized = fe.process_data(csv_path)
df_normalized
```

```python
def get_random_flight(df):
    flight_ids = df['Ac_id'].unique()
    random_flight_id = random.choice(flight_ids)
    flight_data = df[df['Ac_id'] == random_flight_id]
    return random_flight_id, flight_data
```

```python
random_flight_id, actual_flight_data = get_random_flight(df_normalized)
random_flight_id
```

```python
actual_flight_data
```

```python
model = FlightLSTM.load_from_checkpoint('../log/epoch=55-val_loss=0.00.ckpt')
model.to(device)
model.eval()
```

```python
flight_data_normalized = actual_flight_data
wind_conditions = flight_data_normalized[
    [col for col in flight_data_normalized.columns if 'wind_speed' in col or 'wind_dir' in col]].values
print(f"Wind conditions shape: {wind_conditions.shape}")  # Debugging
wind_conditions = torch.FloatTensor(wind_conditions).unsqueeze(0).to(device)
print(f"Wind conditions tensor shape: {wind_conditions.shape}")  # Debugging
with torch.no_grad():
    predicted_path = model(wind_conditions).squeeze(0).cpu().numpy()
```
